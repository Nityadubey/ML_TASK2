(1) Pattern explorer

> Print a right angled tringle of * .

```python
def print_right_triangle(rows):
  for i in range(1, rows + 1):
    print("*" * i)

# Example:
print_right_triangle(5)
```

> Print a number pyramid.



```python
def print_number_pyramid(rows):
  for i in range(1, rows + 1):
    for j in range(rows - i):
      print(" ", end="")
    for k in range(1, i + 1):
      print(k, end="")
    for l in range(i - 1, 0, -1):
      print(l, end="")
    print()

# Example:
print_number_pyramid(5)
```

> Print a checkerboard.





```python
def print_checkerboard(size):

  for i in range(size):
    for j in range(size):
      if (i + j) % 2 == 0:
        print("*", end=" ")
      else:
        print(" ", end=" ")
    print()

# An 8x8 checkerboard:
print_checkerboard(8)
```

(2) Custom calculator.

> Build a calculator that supports.(+,-,*,/,%,**)


> Handle division by zero with try -except.


> Add option for factorial.




```python
import math

def calculator(num1, operator, num2=None):
  if operator == '+':
    if num2 is not None:
      return num1 + num2
    else:
      return "Error: Addition requires two numbers"
  elif operator == '-':
    if num2 is not None:
      return num1 - num2
    else:
      return "Error: Subtraction requires two numbers"
  elif operator == '*':
    if num2 is not None:
      return num1 * num2
    else:
      return "Error: Multiplication requires two numbers"
  elif operator == '/':
    if num2 is not None:
      try:
        return num1 / num2
      except ZeroDivisionError:
        return "Error: Division by zero"
    else:
      return "Error: Division requires two numbers"
  elif operator == '%':
    if num2 is not None:
      try:
        return num1 % num2
      except ZeroDivisionError:
         return "Error: Modulo by zero"
    else:
      return "Error: Modulo requires two numbers"
  elif operator == '**':
    if num2 is not None:
      return num1 ** num2
    else:
      return "Error: Exponentiation requires two numbers"
  elif operator == '!':
    if num2 is None:
      if not isinstance(num1, int) or num1 < 0:
        return "Error: Factorial is only defined for non-negative integers"
      else:
        return math.factorial(num1)
    else:
      return "Error: Factorial is a unary operation and only takes one number"
  else:
    return "Error: Invalid operator"

# Example :
print(f"10 + 5 = {calculator(10, '+', 5)}")
print(f"10 - 5 = {calculator(10, '-', 5)}")
print(f"10 * 5 = {calculator(10, '*', 5)}")
print(f"10 / 5 = {calculator(10, '/', 5)}")
print(f"10 % 3 = {calculator(10, '%', 3)}")
print(f"2 ** 3 = {calculator(2, '**', 3)}")
print(f"Factorial of 5: {calculator(5, '!')}")

print(f"10 / 0 = {calculator(10, '/', 0)}")
print(f"10 ? 5 = {calculator(10, '?', 5)}")
print(f"Factorial of -1: {calculator(-1, '!')}")
print(f"Factorial of 3.5: {calculator(3.5, '!')}")
print(f"Factorial with two numbers: {calculator(5, '!', 2)}")

# Example of division by zero handled with try-except
result_div_zero = calculator(10, '/', 0)
print(f"10 / 0 = {result_div_zero}")

# Example of modulo by zero handled with try-except
result_mod_zero = calculator(10, '%', 0)
print(f"10 % 0 = {result_mod_zero}")
```

(3) Word analyzer

> Take a paragraph as input.


> Count :

*   total word
*   frequency of each word
*   longest word



> Print top 3 most frequent words .









```python
import re
from collections import Counter

def analyze_paragraph(paragraph):
  cleaned_paragraph = re.sub(r'[^\w\s]', '', paragraph).lower()
  words = cleaned_paragraph.split()
  total_words = len(words)
  print(f"Total words: {total_words}")
  word_counts = Counter(words)
  print("\nWord frequency:")
  for word, count in word_counts.items():
    print(f"'{word}': {count}")
  if words:
    longest_word = max(words, key=len)
    print(f"\nLongest word: '{longest_word}'")
  else:
      print("\nNo words found in the paragraph.")
  print("\nTop 3 most frequent words:")
  for word, count in word_counts.most_common(3):
    print(f"'{word}': {count}")

# Example :
input_paragraph = """Nature is the connection between the physical world
surrounding us and the life inside us.Nature is God's most precious and
valuable gift to humans."""

analyze_paragraph(input_paragraph)
```

(i) Nice Pairs
You are given an array a of length n .
A pair (i, j) is called nice if a[i] + a[j] is divisible by k .
Count the number of nice pairs.
# Input
n = 5, k = 3
a = [1, 2, 3, 4]

```python
def count_nice_pairs(a, k):
  n = len(a)
  count = 0
  for i in range(n):
    for j in range(i + 1, n):
      if (a[i] + a[j]) % k == 0:
        count += 1
  return count
n=5
k=3
a=[1,2,3,4]
nice_pairs_count = count_nice_pairs(a, k)
print(f"The number of nice pairs is: {nice_pairs_count}")

# Example with the actual n=5 and a 5-element array
n_example = 5
k_example = 3
a_example = [1, 2, 3, 4, 5]
nice_pairs_count_example = count_nice_pairs(a_example, k_example)
print(f"For a = {a_example} , k = {k_example},the number of nice pairs is: {nice_pairs_count_example}")
```

(ii) Good Subarray
You are given an array of length n .
A subarray is good if the sum of its elements is divisible by its length.
Find the number of good subarrays.
# Input
n = 4
a = [2, 4, 6, 2]
# Output
9

```python
def count_good_subarrays_simple(numbers):
  list_length = len(numbers)
  good_count = 0
  for start_index in range(list_length):
    for end_index in range(start_index, list_length):
      current_subarray = numbers[start_index : end_index + 1]
      subarray_size = len(current_subarray)
      total_sum = sum(current_subarray)
      if subarray_size > 0 and total_sum % subarray_size == 0:
        good_count += 1
  return good_count

my_list = [2, 4, 6, 2]
result_count = count_good_subarrays_simple(my_list)
print(f"For the list {my_list}, the number of good subarrays is: {result_count}")
expected = 9
print(f"The expected output was: {expected}")
print(f"Does our result match the expected output? {result_count == expected}")
```

(iii) Binary String Flips
You are given a binary string. In one operation, you can flip one character ( 0 → 1 or 1 → 0 ).
Find the minimum number of flips required to make the string alternating (like 0101... or 1010... ).
# Input
s = "1001"
# Output
2

```python
def min_flips_to_alternating(s):
  n = len(s)
  flips_to_0101 = 0
  for i in range(n):
    if i % 2 == 0:
      if s[i] == '1':
        flips_to_0101 += 1
    else:
      if s[i] == '0':
        flips_to_0101 += 1
  flips_to_1010 = 0
  for i in range(n):
      if i % 2 == 0:
          if s[i] == '0':
              flips_to_1010 += 1
      else:
          if s[i] == '1':
              flips_to_1010 += 1
  return min(flips_to_0101, flips_to_1010)

s = "1001"

min_flips = min_flips_to_alternating(s)
print(f"For the string '{s}', the minimum number of flips is: {min_flips}")

expected_output = 2
print(f"Expected output: {expected_output}")
print(f"Result matches expected output: {min_flips == expected_output}")

s_example = "0101010101"
min_flips_example = min_flips_to_alternating(s_example)
print(f"\nFor the string '{s_example}', the minimum number of flips is: {min_flips_example}")

s_example_2 = "10101011"
min_flips_example_2 = min_flips_to_alternating(s_example_2)
print(f"For the string '{s_example_2}', the minimum number of flips is: {min_flips_example_2}")
```

(iv) Equalize the Array
You are given an array of integers.
In one operation, you can choose any element and increase or decrease it by 1.
Find the minimum operations to make all elements equal.
# Input
a = [1, 2, 3]
# Output
2

```python
def min_operations_to_equalize(a):
  sorted_a = sorted(a)
  n = len(sorted_a)
  if n % 2 == 1:
    median = sorted_a[n // 2]
  else:
    median = sorted_a[n // 2 - 1]
  operations = 0
  for x in a:
    operations += abs(x - median)
  return operations

a = [1, 2, 3]

min_ops = min_operations_to_equalize(a)
print(f"For the array {a}, the minimum operations to equalize is: {min_ops}")

expected_output = 2
print(f"Expected output: {expected_output}")
print(f"Result matches expected output: {min_ops == expected_output}")

a_example = [1, 1, 2, 2, 3]
min_ops_example = min_operations_to_equalize(a_example)
print(f"\nFor the array {a_example}, the minimum operations to equalize is: {min_ops_example}")

a_example_2 = [5, 1, 4, 2, 8]
min_ops_example_2 = min_operations_to_equalize(a_example_2)
print(f"For the array {a_example_2}, the minimum operations to equalize is: {min_ops_example_2}")
```

(v) Minimum Deletions for Palindrome
You are given a string. In one operation, you can delete one character.
Find the minimum number of deletions to make the string a palindrome.
# Input
s = "aebcbda"
# Output
2

```python
def min_deletions_to_palindrome(s):
  n = len(s)
  dp = [[0 for _ in range(n)] for _ in range(n)]
  for i in range(n):
    dp[i][i] = 1
  for length in range(2, n + 1):
    for i in range(n - length + 1):
      j = i + length - 1
      if s[i] == s[j]:
        if length == 2:
          dp[i][j] = 2
        else:
          dp[i][j] = dp[i + 1][j - 1] + 2
      else:
        dp[i][j] = max(dp[i + 1][j], dp[i][j - 1])
  lps_length = dp[0][n - 1]
  minimum_deletions = n - lps_length

  return minimum_deletions
s = "aebcbda"

min_dels = min_deletions_to_palindrome(s)
print(f"For the string '{s}', the minimum number of deletions to make it a palindrome is: {min_dels}")

expected_output = 2
print(f"Expected output: {expected_output}")
print(f"Result matches expected output: {min_dels == expected_output}")

s_example = "google"
min_dels_example = min_deletions_to_palindrome(s_example)
print(f"\nFor the string '{s_example}', the minimum number of deletions is: {min_dels_example}")

s_example_2 = "racecar"
min_dels_example_2 = min_deletions_to_palindrome(s_example_2)
print(f"For the string '{s_example_2}', the minimum number of deletions is: {min_dels_example_2}")
```
